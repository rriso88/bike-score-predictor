2. Update missing walkscores

```python
walkscore_updated = pd.read_pickle('city_walkscores_updated.pkl')
```

```python
df['city_state'] = df['city']+'_'+df['state']
temps_updated['city_state'] = temps_updated['city']+'_'+temps_updated['state']
walkscore_updated['city_state'] = walkscore_updated['city']+'_'+walkscore_updated['state']
```

```python
df['population'] = np.where(df['population'].isnull(),df['city_state'].map(walkscore_updated.set_index(['city_state'])['population']),df['population'])

```

```python
def update_cols(df, df2, col_update, col_id):
    df[col_update] = np.where(df[col_update].isnull(),df[col_id].map(df2.set_index([col_id])[col_update]),df[col_update])

```

```python
temp_cols = ['percip_inches', 'year_F']
for t in temp_cols:
    update_cols(df, temps_updated, t, 'city_state')
```

```python
walk_cols = ['walk_score', 'transit_score','bike_score','population']
for t in walk_cols:
    update_cols(df, walkscore_updated, t, 'city_state')
```

```python
state_tax = df.groupby(['state','state_tax']).count().reset_index()
state_tax = state_tax.iloc[:,0:2]
state_tax_dict = dict(zip(state_tax.iloc[:,0],state_tax.iloc[:,1]))
state_tax_dict

df['state_tax'] = df['state'].map(state_tax_dict)
```
